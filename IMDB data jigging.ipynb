```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
rating_df = pd.read_csv("resources/title.ratings.tsv", sep='\t')
rating_df.sort_values('averageRating', ascending=False)
```

```python
basics_df = pd.read_csv("resources/title.basic.tsv", sep='\t', low_memory=False)
basics_df
```

```python
basics_df['titleType'].unique()
```

```python
filtered_df = basics_df.loc[basics_df['titleType'] == 'movie']
filtered_df.head(20)
```

```python
merged_data = pd.merge(rating_df, filtered_df, on="tconst", how="left")
merged_data.head(20)
```

```python
clean_data = merged_data.dropna()
```

```python
rating_clean_data = clean_data.sort_values('numVotes', ascending=False)
rating_clean_data.head(20)
```

```python
rating_clean_data1 = rating_clean_data[rating_clean_data['numVotes'] > 200000]
new_rating = rating_clean_data1.sort_values('averageRating', ascending=False)
```

```python
new_rating.reset_index(inplace=True)
```

```python
new_rating.head(20)
```

```python
sorted_rating_df = new_rating[['averageRating', 'numVotes', 'primaryTitle']]
rename_df = sorted_rating_df.rename(columns={'primaryTitle': 'Title', 'averageRating': 'Movie Rating', 'numVotes': 'User Votes'})
```

```python
rename_df.head(50)
```

```python
rename_df.to_csv('imdb_data.csv', index=False)
```
